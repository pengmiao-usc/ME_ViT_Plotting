## 1. Latency per mode

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.array([[0.293116667,1.10208,2.569536667],
        [0.260716667,0.97972,2.169856667],
        [0.068976667,0.2991,0.555136667],
        [0.019186667,0.10155,0.145216667]
])

data_ideal = np.array([[0.27648,1.03716,2.38592], 
                       [0.24576,0.88224,2.00704], 
                       [0.06144,0.2696,0.512], 
                       [0.01536,0.0868,0.13312]
])


fig, ax = plt.subplots(figsize=(10, 4),dpi=100) # use 500 in paper
 
FONT_SIZE=16


sns.set()
sns.set_style("whitegrid")
colors = sns.color_palette("crest")



bar_width = 0.15
cluster_positions = np.arange(4)

labels = ["LP Mode (HLS)", "MSA Mode (HLS)", "MLP Mode (HLS)"]
labels_2 = ["LP Mode (Ideal)", "MSA Mode (Ideal)", "MLP Mode (Ideal)"]

for i in range(3):
    bar_positions = cluster_positions + i * bar_width
    ax.bar(bar_positions, data_ideal[:, i], bar_width, alpha=1, color=colors[2*i], label=labels_2[i])
    ax.bar(bar_positions, data[:, i]-data_ideal[:, i], bar_width, alpha=1, label=labels[i],color=colors[2*i+1],bottom = data_ideal[:, i])


#ax.set_xlabel('Breakdown of Latency per ME-ViT Mode')
#ax.set_xlabel('Models',fontsize=FONT_SIZE)
ax.set_ylabel('Latency (ms)',fontsize=FONT_SIZE)


ax.set_xticks(cluster_positions + bar_width,fontsize=FONT_SIZE)
ax.set_xticklabels(['ViT-B', 'DeiT-B', 'DeiT-S', 'DeiT-T'],fontsize=FONT_SIZE)
ax.set_yticklabels(ax.get_yticks(), fontsize=FONT_SIZE)  # Set the font size of y-axis tick labels

ax.legend(fontsize=14)
plt.show()
```

## 4. Bandwidth per mode

```python


data = np.array([[23.03333333,20.82054842,19.18969957,21.73933487],
                 [23.0125,21.31128038,20.27755102,22.67193997	],
                 [26.825,21.77329377,21.312	,.29220902],
                 [34.45,23.71539171,23.26153846,24.85654379]

])

data_ideal = np.array([[3.566666667,1.70607428,2.391309013,2.189299952],
                       [3.8125,2.005658324,2.781505102,2.542500761	],
                       [5.225,1.640830861,3.14775,3.14775	],
                       [8.05,1.274101382,3.837980769,2.833006425]

])

fig, ax = plt.subplots(figsize=(10, 4),dpi=100)

FONT_SIZE=16


sns.set()
sns.set_style("whitegrid")
colors = sns.color_palette("crest")
colors2 = sns.color_palette("mako_r")


bar_width = 0.1
cluster_positions = np.arange(4)

labels = ["LP Mode (Unoptimized)", "MSA Mode (Unoptimized)", "MLP Mode (Unoptimized)"]
labels_2 = ["LP Mode (Optimized)", "MSA Mode (Optimized)", "MLP Mode (Optimized)"]

for i in range(3):
    bar_positions = cluster_positions + i * bar_width
    ax.bar(bar_positions, data[:, i], bar_width, alpha=1, color=colors[i], label=labels[i])


for i in range(3):

    bar_positions = cluster_positions + (i+3) * bar_width
    ax.bar(bar_positions, data_ideal[:, i], bar_width, alpha=1, label=labels_2[i],color=colors2[i+3])


ax.set_ylabel('Bandwidth (GB/s)',fontsize=FONT_SIZE)


ax.set_xticks(cluster_positions + bar_width,fontsize=FONT_SIZE)
ax.set_xticklabels(['ViT-B', 'DeiT-B', 'DeiT-S', 'DeiT-T'],fontsize=FONT_SIZE)
ax.set_yticklabels(ax.get_yticks().astype('int'),fontsize=FONT_SIZE)  # Set the font size of y-axis tick labels


legend = ax.legend(fontsize=14, bbox_to_anchor=(0.88, -0.15),borderaxespad=0., ncol=2)

#plt.subplots_adjust(right=0.8)  # Adjust the right margin

plt.show()
```
